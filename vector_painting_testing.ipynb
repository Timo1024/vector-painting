```python
import numpy as np
import matplotlib.pyplot as plt
import random
import math
```

```python
vectors = []
iterations = 30

for i in range(iterations):
    x_direction = (random.random()-.5)*1*((.6)**i)
    y_direction = (random.random()-.5)*1*((.6)**i)
    # x_direction = (random.random()-.5)*(iterations-i)/2
    # y_direction = (random.random()-.5)*(iterations-i)/2
    current_vector = [x_direction, y_direction]
    vectors.append(current_vector)

# vectors = [[1, 1], [0, 1], [1, 0]]

vectors = np.array(vectors)
# print(vectors)

origins = []

# start origin
current_origin = np.array([0, 0])

for vector in vectors:
    origins.append(list(current_origin))
    current_origin = current_origin + np.array(vector)

origins = np.array(origins)
# print(origins)

updated_vectors = []

for index, vector in enumerate(vectors):
    updated_vectors.append(np.array(vector) + np.array(origins[index]))

updated_vectors = np.array(updated_vectors)
# print(updated_vectors)
```

```python
def getRadius(start, end):
    x = abs(start[0]-end[0])
    y = abs(start[1]-end[1])
    z = math.sqrt(x*x + y*y)
    return z
```

```python
ax = plt.gca()

for index, vector in enumerate(updated_vectors):
    plt.plot([origins[index][0], vector[0]], [origins[index][1], vector[1]], "black")
    radius = getRadius(origins[index], vector)
    circle = plt.Circle(origins[index], radius, color='r', fill=False)
    ax.add_patch(circle)

# ax.set_xlim([0, 2.2])
# ax.set_ylim([0, 2.2])
ax.axis('equal')
# hi

plt.show()
```
